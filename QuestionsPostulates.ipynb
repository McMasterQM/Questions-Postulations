Which is NOT a postulate of quantum mechanics?
ANS: The wave function must be symmetric with respect to the interchange of all coordinates of one fermion with those of another.
The state of a quantum mechanical system is completely specified by the wave function.
Every observable in classical mechanics corresponds to a linear, Hermitian operator in quantum mechanics.
In any measurement of an observable associated with operator A, only the eigenvalues a which satisfy the following equation can ever be observed: $\hat{A}\Psi = a\Psi$
If a system is in a state described by a normalized wave function Ψ, then the expectation value of the observable corresponding to A is given by: $\langle\hat{A}\rangle=\int_{-\infty}^{\infty}\Psi^{*}\left(\tau\right)\hat{A}\Psi\left(\tau\right)d\tau$
The wave function of a system evolves through time according to the time-dependent Schrödinger equation: $\hat{H}\Psi\left(\vec{r},t\right)=i\hbar\frac{\partial\Psi\left(\vec{r},t\right)}{\partial t}$
